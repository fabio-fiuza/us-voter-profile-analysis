# **Answered Questions**

## PARTE 1 - O PÚBLICO DA PESQUISA

- Qual o perfil da amostra analisada? Verifique a distribuição de idade, gênero, raça, nível educacional e salarial.
- Assumindo que a amostra representa adequadamente a população americana, qual a raça que possui maior poder aquisitivo? E menor?
- Ainda assumindo que temos uma amostra representativa, podemos dizer que nível educacional e poder aquisitivo estão correlacionados?

```python

# visualizando distribuiçào de idade

# OBS: Faixas Etárias Usadas em Políticas Públicas e Estatísticas Demográficas

age_range = nonvoters.ppage.describe()[['min', 'max']]
bins = bins = [age_range['min'], 44, 64, age_range['max']] 
labels = ["Young and Adults\n(18-44)", "Middle-aged\n(45-59)", "Elderly\n(65+)"]
age_group = pd.cut(nonvoters["ppage"], bins=bins, labels=labels, right=False)
age_group = age_group.value_counts(normalize=True).mul(100).sort_index().reset_index()
age_group.columns = ['age', 'proportion']

plt.figure(figsize=(12, 6))

ax = sns.barplot(data=age_group, y='age', x='proportion', color='steelblue', width=0.6)

plt.suptitle('Age Profile of Non-Voters',  fontweight='bold', size=20)
plt.title('Most non-voters are between 22 and 44 years old.', size=15)
plt.xlabel('')
plt.ylabel('Age group', size=12)

for bar, proportion in zip(ax.patches, age_group['proportion']):
    ax.text(bar.get_width() - 0.7, bar.get_y() + bar.get_height() / 2, 
            f'{proportion:.0f}%', color='white', ha='right', va='center', fontsize=12, fontweight=600)
    
plt.xticks([])
plt.yticks(size=10)

ax.spines['left'].set_linewidth(5) 
ax.spines['left'].set_color('black') 
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)


plt.tight_layout()

plt.show()
```

```python
# visualizando distribuição de gênero

gender_group = nonvoters.gender.value_counts(normalize=True).mul(100).reset_index()

plt.figure(figsize=(12, 6))

ax = sns.barplot(data=gender_group, x='gender', y='proportion', hue='gender', palette=['steelblue', 'firebrick'], width=0.5, legend=True)

plt.suptitle('Gender Profile of Non-Voters',  fontweight='bold', size=20)
plt.title('Both groups show a similar contribution', y=1.05,  size=15)
plt.xlabel('')
plt.ylabel('')
plt.legend(title='Gender')

ax.text(x=-0.03, y=25, s=f'{gender_group.proportion.loc[0]:.1f}%', size=14, color='white', fontweight='bold')
ax.text(x=0.95, y=25, s=f'{gender_group.proportion.loc[1]:.1f}%', size=14,  color='white', fontweight='bold')
    
plt.xticks([])
plt.yticks([])

ax.spines['bottom'].set_linewidth(5) 
ax.spines['bottom'].set_color('black') 
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.tight_layout()

plt.show()
```

```python
# visualizando distribuição de raças

race_group = nonvoters.race.value_counts(normalize=True).mul(100).reset_index()

plt.figure(figsize=(12, 6))

ax = sns.barplot(data=race_group, y='race', x='proportion', color='steelblue')

plt.suptitle('Race Profile of Non-Voters',  fontweight='bold', size=20)
plt.title('Most of the population is white.', y=1.05, size=15)
plt.xlabel('')
plt.ylabel('Race', size=12)

for bar, proportion in zip(ax.patches, race_group['proportion']):
    ax.text(bar.get_width() - 0.7, bar.get_y() + bar.get_height() / 2, 
            f'{proportion:.0f}%', color='white', ha='right', va='center', fontsize=12, fontweight=600)
    
plt.xticks([])
plt.yticks(size=10)

ax.spines['left'].set_linewidth(5) 
ax.spines['left'].set_color('black')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

plt.tight_layout()
 
plt.show()
```

```python
# visualizando distribuição de nível educacional

educ_group = nonvoters.educ.value_counts(normalize=True).mul(100).reset_index()

plt.figure(figsize=(12, 6))

ax = sns.barplot(data=educ_group, y='educ', x='proportion', color='steelblue', width=0.6)

plt.suptitle('Educational Profile of Non-voters',  fontweight='bold', size=20)
plt.title('40% have completed college.', y=1.05, size=15)
plt.xlabel('')
plt.ylabel('Educational level ', size=12)

for bar, proportion in zip(ax.patches, educ_group['proportion']):
    ax.text(bar.get_width() - 0.7, bar.get_y() + bar.get_height() / 2, 
            f'{proportion:.0f}%', color='white', ha='right', va='center', fontsize=12, fontweight=600)
    
plt.xticks([])
plt.yticks(size=10)

ax.spines['left'].set_linewidth(5) 
ax.spines['left'].set_color('black')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)

plt.tight_layout()

plt.show()
```

```python
# visualizando distribuição de renda

income_group = nonvoters.income_cat.value_counts(normalize=True).mul(100).reset_index()
income_group['income_cat'] = pd.Categorical(income_group['income_cat'], categories=['Less than $40k', "$40-75k", "$75-125k",  "$125k or more"][::-1], ordered=True)
income_group = income_group.sort_values('income_cat')

income_group = income_group.replace({
    "Less than $40k": "< 40k",
    "$40-75k": "40k-75k",
    "$125k or more": ">= 125k",
    "$75-125k": "75k-125k",
})

plt.figure(figsize=(12, 6))

ax = sns.barplot(data=income_group, y='income_cat', x='proportion', color='steelblue')

plt.suptitle('Income Profile of Non-Voters',  fontweight='bold', size=20)
plt.title('The majority have an annual income of 75k to 125k', y=1.05, size=15)
plt.xlabel('')
plt.ylabel('Income level', size=12)

for bar, proportion in zip(ax.patches, income_group['proportion']):
    ax.text(bar.get_width() - 0.7, bar.get_y() + bar.get_height() / 2, 
            f'{proportion:.0f}%', color='white', ha='right', va='center', fontsize=12, fontweight=600)
    
plt.xticks([])
plt.yticks(size=10)

ax.spines['left'].set_linewidth(5) 
ax.spines['left'].set_color('black')
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)


plt.tight_layout()

plt.show()
```

```python
race_income_table = pd.crosstab(nonvoters['race'], nonvoters['income_cat'])
race_income_table
```

```python
proportions = race_income_table.div(race_income_table.sum(axis=1), axis=0)
proportions
```

```python
highest_income = proportions["$125k or more"].idxmax()
lowest_income = proportions["Less than $40k"].idxmax()

print(f"\nRaça com maior poder aquisitivo: {highest_income}")
print(f"Raça com menor poder aquisitivo: {lowest_income}")
```
